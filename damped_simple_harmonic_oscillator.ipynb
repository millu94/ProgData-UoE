```python

"""
   Damped harmonic oscillator
"""
import math
import matplotlib.pyplot as plt     # Import the plot package as plt

# determine whether damping is over-damped, critically damped, or under-damped,
# return x(t) ie displacement at time t
def evaluateDampedCondition(gamma, omega_0, t):
    a = 1
    if gamma > 2*omega_0:
        condition = "Over-Damped"
        p = math.sqrt((gamma**2)/4 - omega_0**2)
        b = gamma/(2*p)
        x = math.exp(-gamma*t/2)*(a*math.cosh(p*t) + b*math.sinh(p*t))
    
    elif gamma == 2*omega_0:
        condition = "Critically Damped"
        b = gamma/2
        x = math.exp(-gamma*t/2)*(a+b*t)

    else:
        condition = "Under-Damped"
        omega = math.sqrt(omega_0**2 - gamma**2/4)
        b = gamma/2*omega
        x = math.exp(-gamma*t/2)*(a*math.cos(omega*t) + b*math.sin(omega*t))
    
    return x, condition


def main():

    # Take in values for: damping coefficient per unit mass,
    # initial angular velocity,  plot points
    gamma = float(input("Damping Coefficient Per Unit Mass : "))
    print("Gamma : " + str(gamma))
    omega_0 = float(input("Initial Angular Velocity : "))
    print("Omega_0 : " + str(omega_0))
    plotPoints = int(input("Plot Points : "))
    print("Plot Points : " + str(plotPoints))

    t_data = []
    x_data = []
    
    maxTime = (5*math.pi)/omega_0
    print("Total Time : " + str(maxTime))
    increment = maxTime/plotPoints

    for i in range(0, plotPoints): 
        t = increment*i
        x, condition = evaluateDampedCondition(gamma, omega_0, t)
        t_data.append(t)
        x_data.append(x)
        

    #print(str(t_data))
    #print(str(x_data))

    # Plot the trajectory 
    plt.plot(t_data,x_data,"b")          
    plt.title(f"{condition} Simple Harmonic Oscillator")
    # plt.title("Simple Harmonic Oscillator")
    
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.show()                           

    

main()
```
